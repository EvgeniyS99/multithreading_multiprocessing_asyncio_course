### generator examples

```python
def gen():
    yield 1
    yield 2
g = gen()
# print(next(g))
# print(next(g))
# print(next(g))
```

```python
g
```

```python
print(next(g))
print(next(g))
print(next(g))
```

```python
def gen():
    print('HI inside gen')
    x = yield 1
    print("result inside gen: x =", x)
    return x

g = gen()
# print(next(g))
# print(next(g))
print('yield', g.send(None))
try:
    res = g.send(9)
except StopIteration as exc:
    print(exc.value)
```

```python
def child():
    print("child: start")
    yield "child -> 1"
    print("child: before return")
    return 10

def parent():
    print("parent: start")
    x = yield from child()
    print("parent: child returned", x)
    yield f"parent got {x}"
    return x + 1
```

```python
gen_p = parent()
```

```python
gen_p.send(1)
```

```python
gen_p.send(None)
```

```python
gen_p.send(None)
```

## Nested coroutines

```python
import asyncio
```

```python
async def child():
    print('start child')
    await asyncio.sleep(0.1)
    #return 89

async def main():
    print('start')
    await child()
    print('next')
    await child()
    print('one more next')
    return 8
```

```python
coro = main()
coro
```

```python
r = coro.send(None)
r
```

```python
coro.send(None)
```

```python
coro.send(None)
```

```python
# await something <-> yield from something.__await__() 
```

```python
class MyAwaitable:
    def __await__(self):
        print('in __await__')
        yield 7
        return 1
```

```python
async def main():
    print('started')
    x = await MyAwaitable()
    print(x)
```

```python
main_cor = main()
res_from_awaitable = main_cor.send(None)
print(res_from_awaitable)
```

```python
main_cor.send(None)
```
